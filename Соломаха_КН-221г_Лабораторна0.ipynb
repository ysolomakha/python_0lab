````python
import math

def sum_of_squares(N):
    
    if not isinstance(N, (int)):
        print("Помилка: Введіть числове значення для N")
        return
    
    if N <= 0:
        print("N має бути додатнім числом")
        return

    if N % 4 != 1:
        print(f"Число {N} не можна представити квадратами цілих чисел")
        return

    max_value = int(math.sqrt(N))
    for a in range(1, max_value + 1):
        b_squared = N - a**2
        b = int(math.sqrt(b_squared))
        if b**2 == b_squared:
            print(f"Знайдено: a = {a}, b = {b}, {a}^2 + {b}^2 = {N}")
            return

    print(f"Число {N} не можна представити квадратами цілих чисел")
````
